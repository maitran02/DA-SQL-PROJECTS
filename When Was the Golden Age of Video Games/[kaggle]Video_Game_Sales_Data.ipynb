```python
%%sql

DROP TABLE IF EXISTS game_sales_data;
CREATE TABLE game_sales_data(
    rank NUMERIC,
    name VARCHAR(500),
    platform VARCHAR(10),
    publisher VARCHAR(100),
    developer VARCHAR(500),
    critic_score NUMERIC,
    user_score NUMERIC,
    total_shipped NUMERIC,
    year INT
    );

COPY game_sales_data
FROM 'D:\OneDrive - ftu.edu.vn (1)\WORKING SPACE\DATA ANALYSIS\DATACAMP\Project\When Was the Golden Age of Video Games\game_sales_data.csv'
DELIMITER ','
csv header encoding 'windows-1251';
```

```python
%%sql
DROP VIEW IF EXISTS game_sales;
CREATE VIEW game_sales AS
SELECT name, platform, publisher, developer, total_shipped, year
FROM game_sales_data;

SELECT * FROM game_sales
WHERE name = 'Minecraft'
LIMIT 5;
```

```python
%%sql
DROP VIEW IF EXISTS reviews;

CREATE VIEW reviews AS
SELECT name, critic_score, user_score
FROM game_sales_data;

SELECT *
FROM reviews
WHERE name = 'Minecraft'
```

```python
%%sql
-- TOP 10 BEST SELLING GAMES
SELECT name, total_shipped FROM game_sales
ORDER BY total_shipped DESC
LIMIT 10;
```

```python
%%sql
SELECT *
FROM game_sales
LEFT JOIN reviews
USING (name)
LIMIT 19
-- WHERE critic_score IS NULL AND user_score IS NULL
```

```python
%%sql
-- Select a count of the number of games where both critic_score and user_score are null
SELECT (SELECT COUNT(name)
FROM game_sales
LEFT JOIN reviews
USING (name)
WHERE critic_score IS NULL AND user_score IS NULL
) AS count_score, COUNT(*)
```
